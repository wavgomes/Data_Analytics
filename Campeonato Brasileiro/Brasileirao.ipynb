```
def Vencedor(a,b):
  if a == b:
    return 'Mandante'
  elif b == 'Empate':
    return 'Empate'
  else:
    return 'Visitante'
```

```
brasileirao_10_vitorias = brasileirao_10[brasileirao_10['Estado Vencedor'] != 'Empate']
```

```
brasileirao_10_vitorias = brasileirao_10_vitorias.merge(estados, how='inner', left_on='Estado Vencedor', right_on='uf')
```

```
brasileirao_10_vitorias.drop(columns=['codigo_uf'],inplace=True)
brasileirao_10_vitorias.rename(columns={'nome':'Estado','latitude':'Latitude','longitude':'Longitude','uf':'UF'}, inplace=True)
```

```
brasileirao_10_vitorias['Time Vencedor'] = brasileirao_10_vitorias.apply(lambda x: x['Mandante'] if x['Gols Mandante'] > x['Gols Visitante'] else x['Visitante'],axis=1)
```

```
for nome_time in brasileirao_10_vitorias['Time Vencedor']:
  if nome_time == nome_time.upper():
    brasileirao_10_vitorias.loc[brasileirao_10_vitorias['Time Vencedor'] == nome_time,'Time Vencedor'] = nome_time.capitalize()
```

```
vitorias_por_ano = brasileirao_10_vitorias.groupby(['Time Vencedor','Ano'],as_index=False)['Placar'].count()
```

```
vitorias_aux = vitorias_por_ano.groupby('Ano',as_index=False)['Placar'].max()
```

```
vitorias_aux.rename(columns={'Placar':'Qtd Max Vitórias Ano'}, inplace=True)
```

```
vitorias_por_ano = vitorias_por_ano.merge(vitorias_aux,how='inner',left_on='Ano',right_on='Ano')
```

```
ganhadores = dict()
for ano in sorted(vitorias_por_ano['Ano'].unique()):
  for time in vitorias_por_ano[vitorias_por_ano['Ano']==ano]['Time Vencedor']:
    vitoria = vitorias_por_ano[(vitorias_por_ano['Time Vencedor']==time)&(vitorias_por_ano['Ano']==ano)]['Placar'].values[0]
    total_vitoria = vitorias_por_ano[(vitorias_por_ano['Time Vencedor']==time)&(vitorias_por_ano['Ano']==ano)]['Qtd Max Vitórias Ano'].values[0]

    if vitoria == total_vitoria:
      #print(time, ano, vitoria, total_vitoria)
      if ano not in ganhadores.keys():
        ganhadores.update({ano:[time,vitoria]})
      else:
        aux = ganhadores[ano][0] + ' & ' + time
        aux1 = ganhadores[ano][1]
        ganhadores.pop(ano)
        ganhadores.update({ano:[aux,aux1]})
```

```
ganhadores_10 = pd.DataFrame.from_dict(ganhadores,orient='index',columns=['Time','Qtd Vitórias']).reset_index()
ganhadores_10.rename(columns={'index':'Ano'},inplace=True)
```

```
cores = dict()
times = ganhadores_10['Time'].unique()
for i,v in enumerate(times):
  for ii,vv in enumerate(px.colors.sequential.Rainbow):
    if i == ii:
      cores.update({v:vv})
```

```
cores_times = pd.DataFrame.from_dict(cores, orient='index', columns=['Cores']).reset_index()
cores_times.rename(columns={'index':'Time'},inplace=True)
ganhadores_10 = ganhadores_10.merge(cores_times,how='inner',left_on='Time',right_on='Time').sort_values('Ano')
```

```
barras = []
for cor, ganhadores_10_aux in ganhadores_10.groupby('Cores'):
  
  barras.append(go.Bar(
        x=ganhadores_10_aux['Ano'],
        y=ganhadores_10_aux['Qtd Vitórias'],
        name=ganhadores_10_aux['Time'].values[0],
        marker=({'color':cor})
    ))
fig = go.Figure(data=barras)
for template in [#"plotly","plotly_white",
                 "plotly_dark"]:
  #fig = go.FigureWidget(data=barras)
  fig.update_layout(template=template)
  fig.show()
```

```
fig = go.Figure()
fig.add_trace(go.Table(header=dict(values=list('<b>' + ganhadores_10.drop('Cores',axis=1).columns + '</b>'),
                                   align=['left','center'],
                                   font_size=12),
                       cells=dict(values=[ganhadores_10['Ano'],ganhadores_10['Time'],ganhadores_10['Qtd Vitórias']],
                                  align=['left','center'])))
fig.update_layout(template='plotly_dark')
fig.show()
```

```
placares_count = dict()

for placar in brasileirao_10['Placar']:
  if placar not in placares_count:
    placares_count[placar] = 1
  else:
    placares_count[placar] += 1
```

```
placares = pd.DataFrame.from_dict(placares_count,orient='index',columns=['Qtd Jogos']).reset_index()
placares.rename(columns={'index':'Resultado'},inplace=True)
placares = placares.sort_values('Qtd Jogos', ascending=False).reset_index(drop=True)
```

```
fig = go.Figure()
fig.add_trace(go.Pie(labels=placares['Resultado'].head(10), 
                     values=placares['Qtd Jogos'].head(10), 
                     textinfo='value+percent',
                     pull=[0.1]))
for template in ['plotly_dark']:
  fig.update_layout(template=template)
  fig.show()
```

```
placares_count_distinct = dict()

for placar in brasileirao_10['Placar']:
  x = placar[::-1]
  if (placar not in placares_count_distinct) & (x not in placares_count_distinct):
    placares_count_distinct[placar] = 1
  elif (placar in placares_count_distinct) & (x not in placares_count_distinct):
    placares_count_distinct[placar] += 1
  else:
    placares_count_distinct[x] += 1
```

```
placares_distinct = pd.DataFrame.from_dict(placares_count_distinct,orient='index',columns=['Qtd de Jogos']).reset_index()
placares_distinct.rename(columns={'index':'Resultado'},inplace=True)
placares_distinct = placares_distinct.sort_values('Qtd de Jogos', ascending=False).reset_index(drop=True)
```

```
fig = go.Figure()
fig.add_trace(go.Pie(labels=placares_distinct['Resultado'].head(10),
                     values=placares_distinct['Qtd de Jogos'].head(10), 
                     textinfo='value+percent',
                     pull = [0.1]))
fig.update_layout(template='plotly_dark')
fig.show()
```
